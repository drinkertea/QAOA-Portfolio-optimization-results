```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

iters = [
    1,
    5,
    10,
    50,
    100,
]

better = [
    ('ibmq_belem (real)', 'drive/MyDrive/run2304/results_ibmq_belem_{}_AAPL_MSFT_AMZN_TSLA_GOOGL.csv'),
    ('ibmq_belem (simulator same noise)', 'drive/MyDrive/run0505/results_ibmq_belem_{}_AAPL_MSFT_AMZN_TSLA_GOOGL.csv'),
    ('simulator no noise', 'drive/MyDrive/run0505sim/results_ibmq_belem_{}_AAPL_MSFT_AMZN_TSLA_GOOGL.csv'),
]
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 5)

for i in range(len(better)):
    print(better[i][0])
    over_iters = {}
    for it in iters:
        df = pd.read_csv(better[i][1].format(it))
        selection_proc = np.sum(df['selection'] == np.int64('01001')) / df.shape[0] * 100
        contains_proc = np.sum(df['01001'] > 0) / df.shape[0] * 100
        over_iters[str(it)] = df['01001']
        print(selection_proc, contains_proc)
    df = pd.DataFrame(over_iters)
    boxplot = sns.boxplot(x="variable", y="value", data=pd.melt(df))
    boxplot.axes.set_title("{}".format(better[i][0]), fontsize=16)
    boxplot.set_xlabel("Iterations", fontsize=14)
    boxplot.set_ylabel("Amplitude", fontsize=14)
    plt.show()
```
